```python
import geopandas as gpd
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def model_const_source_1order_sink(S,k,m0):
    t = np.linspace(0,25,101)
    m = m0*np.exp(-k*t) + S/k*(1-np.exp(-k*t))
    return t, m
```

```python
#Case 1: S=0
S=0
k = 0.2
m0 = 0.5

t,m = model_const_source_1order_sink(S,k,m0)
plt.figure(figsize=(5,5))
plt.plot(t,m,color='k',linewidth=2)
plt.plot([1/k,1/k],[0,0.5],linestyle='--',color='tab:blue',linewidth=1)
plt.plot([0,25],[m0*np.exp(-1),m0*np.exp(-1)],linestyle='--',color='tab:blue',linewidth=1)
plt.xticks([0,1/k,2/k,3/k,4/k,5/k],['0','1/k','2/k','3/k','4/k','5/k'],fontsize=12)
plt.yticks([0,0.5],['0','m(0)'],fontsize=16)
plt.ylim([0,0.5])
plt.xlim([0,25])
plt.xlabel('t',fontsize=16)
plt.ylabel('m',fontsize=16)
```

```python
#Case 2: S< k x m0
S=0.05
k = 0.2
m0 = 0.5

t = np.linspace(0,25,101)
m = m0*np.exp(-k*t) + S/k*(1-np.exp(-k*t))

plt.figure(figsize=(5,5))
plt.plot(t,m,color='k',linewidth=2)
plt.plot([0,25],[S/k,S/k],linestyle='--',color='tab:orange',linewidth=1)
plt.xticks([0,1/k,2/k,3/k,4/k,5/k],['0','1/k','2/k','3/k','4/k','5/k'],fontsize=12)
plt.yticks([0,0.25,0.5],['0','S/k','m(0)'],fontsize=16)
plt.ylim([0,0.5])
plt.xlim([0,25])
plt.xlabel('t',fontsize=16)
plt.ylabel('m',fontsize=16)
```

```python
#Case 3: S> k x m0
S=0.2
k = 0.2
m0 = 0.5

t = np.linspace(0,25,101)
m = m0*np.exp(-k*t) + S/k*(1-np.exp(-k*t))

plt.figure(figsize=(5,5))
plt.plot(t,m,color='k',linewidth=2)
plt.plot([0,25],[S/k,S/k],linestyle='--',color='tab:orange',linewidth=1)
plt.xticks([0,1/k,2/k,3/k,4/k,5/k],['0','1/k','2/k','3/k','4/k','5/k'],fontsize=12)
plt.yticks([0,0.5,1],['0','m(0)','S/k'],fontsize=16)
#plt.ylim([0,0.5])
plt.xlim([0,25])
plt.xlabel('t',fontsize=16)
plt.ylabel('m',fontsize=16)
```
